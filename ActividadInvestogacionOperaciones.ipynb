```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Definir los datos
df = pd.DataFrame( {
    'MES': ['13/10/2023', '16/10/2023', '17/10/2023', '18/10/2023', '19/10/2023', '20/10/2023', '23/10/2023'],
    'APERTURA': [53.88, 54.07, 53.79, 53.79, 53.76, 53.39, 52.82],
    'PRONOSTICO': [None, 53.975, 53.93, 53.79, 53.775, 53.575, 53.105],
    'ERROR_ABSOLUTO': [None, 0.095, 0.14, 0, 0.015, 0.185, 0.285],
    'ERROR_AB5%': [None, 0.175698169,  0.260271426,  0,  0.027901786,  0.346506836,  0.539568345],
    'MAD': [None, 0.0475, 0.078333333, 0.05875, 0.05, 0.0725, 0.102857143],
    'DCM': [None, 0.0045125, 0.018408333, 0.01380625, 0.0125, 0.0315375, 0.074057143],
    'Limite_superior': [None, 54.10666328, 54.19592753, 54.0203, 53.99413466, 53.92307249, 53.63838346],
    'Limite_inferior': [None, 53.84333672, 53.66407247, 53.5597, 53.55586534, 53.22692751, 52.57161654]
} )

# Convertir la columna MES en tipo fecha
df['MES'] = pd.to_datetime(df['MES'], format='%d/%m/%Y')
```

```python
ymin = df['APERTURA'].min()-1
ymax = df['APERTURA'].max()+1
```

```python
# Imprimir el DataFrame
df
```

```python
def g(columnas, titulo):
  plt.figure(figsize=(10, 6))
  for columna in columnas:
    plt.plot(df['MES'], df[columna], label=columna.capitalize())
  plt.ylim(ymin,ymax)
  plt.xlabel('Mes')
  plt.ylabel('Valor')
  plt.title(titulo)
  plt.legend()
  plt.show()
```

```python
# Graficar Mes, Pronostico, Apertura, y Error Absoluto
g(['PRONOSTICO', 'APERTURA', 'ERROR_ABSOLUTO'], 'Gr√°fico de Mes, Pronostico, Apertura y Error Absoluto')
```

```python
# Graficar Pronostico, Mes y MAD
g(['PRONOSTICO', 'MAD'], 'Gr√°fico de Pronostico, Mes y MAD')
```

```python
# Graficar Pronostico, Apertura DCM, Mes
g(['PRONOSTICO','DCM'], 'Gr√°fico de Pronostico, Apertura DCM, y Mes')
```

```python
# Graficar Mes, Pronostico con respecto a los l√≠mites superior e inferior del pronostico
plt.figure(figsize=(10, 6))
plt.plot(df['MES'], df['PRONOSTICO'], label='Pronostico')
plt.plot(df['MES'], df['Limite_superior'], label='Limite superior')
plt.plot(df['MES'], df['Limite_inferior'], label='Limite inferior')
plt.xlabel('Mes')
plt.ylabel('Valor')
plt.title('Gr√°fico de Mes, Pronostico, L√≠mite superior e inferior')
plt.legend()
plt.show()
```
